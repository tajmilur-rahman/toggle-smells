```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/govardhanrathamsetty/Downloads/1000_comp_matching.csv')
df
```

```python
df['tog_id'] = range(1, len(df)+1)
df
```

```python
comp = pd.read_csv('/Users/govardhanrathamsetty/Downloads/components.csv')
comp
```

```python
path_items=[]
tog_map=[]
path_tog_match =[]

for i in df.file_path:
        
        parts = i.split('/')

        cmpts = parts[8:]
        
        path_items.append(cmpts)
```

```python
x=[]

for i, j in df.iteritems():
        if i == 'file_path':
            parts = j.str.split('/')
            df['cmpts'] = parts.apply(lambda x: x[8:11])

df['matched'] = None

for i, row1 in df.iterrows():
    for _, row2 in comp.iterrows():
            for value in row1['cmpts']:
                if value in row2['name']:           
                    df.at[i, 'matched'] = row2['name']
                break
df[df.matched.notna()]
```
